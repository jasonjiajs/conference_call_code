# Part 2: Get the firm names of conference calls with matches

```python
xls_compiled_path = Path(r"C:\Users\jasonjia\Dropbox\Projects\conference_call\output\12_summary_statistics\xls_compiled_01.1_xls_1_and_2.csv")
unique_conference_calls_with_keywords_path = Path(r"C:\Users\jasonjia\Dropbox\Projects\conference_call\output\12_summary_statistics\unique_conference_calls_with_keywords_all_columns.csv")
```

```python
xls_compiled = pd.read_csv(xls_compiled_path, low_memory=False)
xls_compiled = xls_compiled[['Report #', 'Title', 'Subtitle']]
xls_compiled = xls_compiled.dropna(subset=['Report #'])
xls_compiled = xls_compiled.astype({'Report #': 'int64'})
xls_compiled
```

```python
unique_conference_calls_with_keywords = pd.read_csv(unique_conference_calls_with_keywords_path)
unique_conference_calls_with_keywords = unique_conference_calls_with_keywords[['Report', 'gvkey', 'Title']]
unique_conference_calls_with_keywords = unique_conference_calls_with_keywords.rename(columns = {'Report': 'Report #', 'Title': "Title_v2"})
unique_conference_calls_with_keywords
```

```python
# Merge
final = unique_conference_calls_with_keywords.merge(xls_compiled, how='left', on = "Report #")
final
```

```python
# Check if there are any non-matches - none. Good!
final[final['Title'].isna()]
```

```python
# Identify duplicate report #s
duplicateRows = final[final.duplicated(['Report #'])]
duplicateRows
```

```python
# Do a few checks - confirm that the titles are the same.
final[final['Report #'] == 8720951]
```

```python
# Remove duplicated report #s - should have no impact.
first = final.groupby('Report #').first().reset_index()
first
```

```python
# Check # of NaNs in Title_v2 - confirm that there's 0
first.groupby('Title_v2').first().reset_index()['Title_v2'].isna().sum()
```

```python
# Filter by Title_v2 since there's no NaNs in Title_v2, whereas there are NaNs in Title
second = first.groupby('Title_v2').first().reset_index()
second
```

```python
# There are clearly differences in the titles. I trust title_v2 more, since it's been manually looked at.
second[second['Title_v2'] != second ['Title']]
```

```python
# Replace a with b for all strings in a df column at all occurences
df['Title_v2_truncated'] = df['Title_v2'].str.replace(r'^.*- event .+ of ','', regex = True, case=False)
df['Title_v2_truncated'] = df['Title_v2_truncated'].str.replace(r'conference call.*','', regex = True, case=False)
df['Title_v2_truncated'] = df['Title_v2_truncated'].str.replace(r' -.+ transcript.*$','', regex = True, case=False)
df['Title_v2_truncated'] = df['Title_v2_truncated'].str.replace(r' - preliminary .*$','', regex = True, case=False)
df['Title_v2_truncated'] = df['Title_v2_truncated'].str.replace(r' confere.*$','', regex = True, case=False)
df['Title_v2_truncated'] = df['Title_v2_truncated'].str.replace(r' - Q\d \d\d\d\d .*$','', regex = True, case=False)
```
